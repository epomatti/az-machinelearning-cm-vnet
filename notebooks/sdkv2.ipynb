```python
from azure.ai.ml import MLClient
from azure.ai.ml.entities import Data
from azure.ai.ml.constants import AssetTypes
from azure.identity import DefaultAzureCredential

# 1. Connect to Workspace
ml_client = MLClient.from_config(credential=DefaultAzureCredential())

# 2. Define the Data Asset (pointing to your blob path)
# Note: 'blobs' is your datastore name
my_data = Data(
    path="azureml://datastores/test/paths/contacts.csv",
    type=AssetTypes.URI_FILE,
    version="latest",
    description="Contacts CSV file",
    name="contacts_dataset"
)

# 3. Create/Register it
ml_client.data.create_or_update(my_data)

data_asset = ml_client.data.get(name="contacts_dataset", version="latest")
print(
    f"Data asset already exists. Name: {my_data.name}, version: {my_data.version}"
)
```
